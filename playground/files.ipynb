```python
import os
from pathlib import Path
import csv
```

```python
# define paths

DATA_ROOT_DIR = "../data"
data_path = Path(DATA_ROOT_DIR)
lists_folder = data_path / "lists/"
music_folder = data_path / "music/"
```

```python
# lists

all_artist_list_file = lists_folder / "artist_names.txt"
training_artist_list_file = lists_folder / "training_artist_names.csv"
```

```python
all_artists = []
with open(all_artist_list_file, "r") as r:
    lines = r.readlines()
    for line in lines:
        all_artists.append(line.replace("\n",""))
all_artists = all_artists[1:]
```

```python
print(all_artists[1:])
print(f"Total artists: {len(all_artists)}")
```

```python
top_artists = []
with open(training_artist_list_file, "r") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        top_artists.append(row[0])
print(top_artists)
print(f"Total artists: {len(top_artists)}")
```

```python
for artist in all_artists:
    artist_folder = music_folder / f"{artist}/"
    midis = list(artist_folder.glob("*.mid"))
    print(f"{artist} has {len(midis)} song(s)")
    print("")
```

```python
training_artists = []
for artist in top_artists:
    if artist in all_artists:
        training_artists.append(artist)
```

```python
print(training_artists)
print(f"Total artists: {len(all_artists)}")
print(f"Top artists: {len(top_artists)}")
print(f"Training artists: {len(training_artists)}")
```

```python
midi_count = 0
for artist in training_artists:
    artist_folder = music_folder / f"{artist}/"
    midis = list(artist_folder.glob("*.mid"))
    midi_count += len(midis)
    print(f"{artist} has {len(midis)} song(s)")
print("Total MIDI count", midi_count)
```
